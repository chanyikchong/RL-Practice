## Fundamentals <a id='fundamentals'></a>

### Key Concepts
- **State (s)**: observation of the environment
- **Action (a)**: what the agent does
- **Reward (r)**: scalar feedback signal
- **Policy (pi)**: mapping from states to actions
- **Value function V(s)**: expected cumulative reward from state s
- **Q-function Q(s,a)**: expected cumulative reward from state s, taking action a
- **Discount factor (gamma)**: how much to weight future vs immediate rewards

### The Bellman Equation
$$V(s) = \max_a \left[ r(s,a) + \gamma V(s') \right]$$

## Q-Learning <a id='q-learning'></a>

### Update Rule
$$Q(s,a) \leftarrow Q(s,a) + \alpha \left[ r + \gamma \max_{a'} Q(s',a') - Q(s,a) \right]$$

### Key Innovations

### Loss Function
$$L(\theta) = \mathbb{E}\left[ \left( r + \gamma \max_{a'} Q(s', a'; \theta^-) - Q(s, a; \theta) \right)^2 \right]$$

### Clipped Objective
$$L^{CLIP}(\theta) = \mathbb{E}\left[ \min\left( r_t(\theta) A_t, \text{clip}(r_t(\theta), 1-\epsilon, 1+\epsilon) A_t \right) \right]$$

### GAE (Generalized Advantage Estimation)
